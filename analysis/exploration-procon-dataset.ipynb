### Checking for null and ditribution for each of the Interesting fields

```python
def check_null_and_distribution(field):    
    print(f'Check null for {field}:\n\n')

    null_total_in_field = len(procon_df[procon_df[field].isna()].index)
    print(f'TOTAL NULL: {null_total_in_field}')
    
    field_len = len(list(procon_df[field].unique()))
    print(f'TOTAL of unique values in field: {field_len}')
    
    dist_df = procon_df \
        .groupby([field]) \
        .agg({'TrimestreAtendimento': 'count'}) \
        .sort_values(by='TrimestreAtendimento', ascending=False)
    print(dist_df.head())
    if field_len > 10:
        print(dist_df.tail())
        
    print('\n\n')
```

```python
for field in ['DataAtendimento', 'AnoAtendimento', 'MesAtendimento', 'Regiao', 'UF', 'DescricaoTipoAtendimento',
              'GrupoAssunto', 'GrupoProblema', 'SexoConsumidor', 'FaixaEtariaConsumidor', 'RazaoSocialSindec', 
              'NomeFantasiaSindec', 'CNPJ', 'DescricaoCNAEPrincipal', 'CEPConsumidor']:
    check_null_and_distribution(field)
```
